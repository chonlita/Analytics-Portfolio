# Import Data

```python
import pandas as pd
import numpy as np

df = pd.read_csv('D3 (Assessment 2).csv')
df.info()
```

# 1. Preprocessing

## Valuecounts

```python
categorical_data = ["race","age","gender","medical_specialty","metformin", "repaglinide", "nateglinide", "chlorpropamide",
                    "glimepiride", "acetohexamide", "glipizide","glyburide", "tolbutamide", "insulin"]

for i in categorical_data:
    print(df[i].value_counts())
    print("--------------")
```

## Numerical correlation - no multicollinearity

```python
import seaborn as sns
import matplotlib.pyplot as plt

df2 = df.select_dtypes(include=[np.number])
corr_matrix = df2.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', annot_kws={"size":7})
plt.show()
```

## Dropping Unique Identifier, fillna, replace invalid

```python
#Dropping variables
#Drop IDs
df.drop(["admission_type_id", "discharge_disposition_id", "admission_source_id"], axis ='columns', inplace=True)

#Drop the large invalid
df.drop(["medical_specialty"], axis ='columns', inplace=True)

#Drop the large no
df.drop(["repaglinide", "nateglinide","glimepiride","acetohexamide","tolbutamide","chlorpropamide","glyburide"], axis ='columns', inplace=True)

#Fillna
df['age'].fillna(df['age'].mode()[0], inplace=True)
df['race'].fillna(df['race'].mode()[0], inplace=True)
mask = df['gender'] == "Unknown/Invalid"
df.loc[mask, 'gender'] = np.nan
df['gender'].fillna(df['gender'].mode()[0], inplace=True)
```

```python
df.info()
```

## Data Partition

```python
df = pd.get_dummies(df)
```

```python
y = df['readmitted']
X = df.drop(['readmitted'], axis=1)
X_mat = X.to_numpy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# setting random state
rs = 10

# 70:30 split
X_train, X_test, y_train, y_test = train_test_split(X_mat, y, test_size=0.3, stratify=y, random_state=rs)
```

```python
df.info()
```

```python
df.info
```

# 2. Build a decision tree (Default setting)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score

rs = 10
# simple decision tree training
model = DecisionTreeClassifier(random_state=rs)
model.fit(X_train, y_train)

# print model parameters
print(model.get_params(deep = True))
```

## 2a. Check train and test accuracy

```python
print("Train accuracy:", model.score(X_train, y_train))
print("Test accuracy:", model.score(X_test, y_test))
```

```python
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

## 2b/c. Size of Tree and who decides the first split

```python
tree_size = model.tree_.node_count
print("The size of the tree (number of nodes) is:", tree_size)
```

```python
first_split_variable = model.tree_.feature

# Display the first split variable
print("The variable used for the first split is:",first_split_variable)
```

```python
print(X.columns[6])
```

## 2d. Feature importance

```python
import numpy as np
# grab feature importances from the model and feature name from the original X
importances = model.feature_importances_
feature_names = X.columns
# sort them out in descending order
indices = np.argsort(importances)
indices = np.flip(indices, axis=0)
# limit to 20 features, you can leave this out to print out everything
indices = indices[:20]
for i in indices:
    print(feature_names[i], ':', importances[i])
```

## 2e. Parameters used

The following hyperparameters are specified before training:

criterion: The splitting criterion used to split the data at each node.

max_depth: The maximum depth of the tree.

min_samples_leaf: The minimum number of samples required in a leaf node.

min_samples_split: The minimum number of samples required to split a node.

splitter: The strategy used to split nodes.

---
The following parameters are learned during training:

ccp_alpha: The complexity parameter used to prune the tree.

class_weight: The weights given to different classes.

max_features: The maximum number of features used to split a node.

max_leaf_nodes: The maximum number of leaf nodes in the tree.

min_impurity_decrease: The minimum decrease in impurity required to split a node.

min_weight_fraction_leaf: The minimum weight fraction required in a leaf node.

random_state: The random state used to split nodes.

# 3. Decision tree tuned with GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
# grid search CV
params = {'criterion': ['gini', 'entropy'],
          'max_depth': range(1, 16),
          'min_samples_leaf': range(0, 25, 5)[1:]}

cv_1 = GridSearchCV(param_grid=params, estimator=DecisionTreeClassifier(random_state=rs),return_train_score=True, cv=10)
cv_1.fit(X_train, y_train)

result_set = cv_1.cv_results_
print(result_set)
```

```python
print(cv_1.best_params_)
```

```python
import matplotlib.pyplot as plt

result_set['params']
dd = pd.DataFrame(result_set['params'])
index_ = list(dd.index[(dd['criterion']=='entropy') & (dd['min_samples_leaf']==5)])

train_result = result_set['mean_train_score']
test_result = result_set['mean_test_score']

max_depth_train = []
max_depth_test = []


for i in range(len(index_)):
    max_depth_train.append(train_result[index_[i]])
    max_depth_test.append(test_result[index_[i]])
plt.plot(range(1, len(max_depth_train)+1), max_depth_train, 'b', range(1,len(max_depth_test)+1), max_depth_test, 'r')
plt.xlabel('Maximum depth (model complexity)\nBlue = training acc. Red = test acc.')
plt.xticks(np.arange(1, len(max_depth_train)+1, 1))
plt.ylabel('score')
plt.show()
```

## ROC Curve of Optimal model

```python
dt_cv_best = cv_1.best_estimator_
# probability prediction from decision tree
y_pred_proba_dt = dt_cv_best.predict_proba(X_test)

print("Probability produced by decision tree for each class vs actual prediction on TargetB (0 = non-donor, 1 = donor). You should be able to see the default threshold of 0.5.")
print("(Probs on zero)\t(probs on one)\t(prediction made)")
# print top 10
for i in range(20):
    print(y_pred_proba_dt[i][0], '\t', y_pred_proba_dt[i][1], '\t', y_pred[i])
```

```python
from sklearn.metrics import roc_auc_score

y_pred_proba_dt = model.predict_proba(X_test)
y_pred_proba_dt_cv = dt_cv_best.predict_proba(X_test)

roc_index_dt = roc_auc_score(y_test, y_pred_proba_dt[:, 1])
roc_index_dt_cv = roc_auc_score(y_test, y_pred_proba_dt_cv[:, 1])

print("ROC index on test for DT_default:", roc_index_dt)
print("ROC index on test for DT_optimal:", roc_index_dt_cv)
```

```python
from sklearn.metrics import roc_curve

fpr_dt, tpr_dt, thresholds_dt = roc_curve(y_test, y_pred_proba_dt[:,1])
fpr_dt_cv, tpr_dt_cv, thresholds_dt_cv = roc_curve(y_test, y_pred_proba_dt_cv[:,1])
```

```python
import matplotlib.pyplot as plt

plt.plot(fpr_dt, tpr_dt, label='ROC Curve for DT_default {:.3f}'.format(roc_index_dt), color='red', lw=0.5)
plt.plot(fpr_dt_cv, tpr_dt_cv, label='ROC Curve for DT_optimal {:.3f}'.format(roc_index_dt_cv), color='blue', lw=0.5)

# plt.plot(fpr[2], tpr[2], color='darkorange',
#          lw=lw, label='ROC curve (area = %0.2f)' % roc_auc[2])
plt.plot([0, 1], [0, 1], color='navy', lw=0.5, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```
